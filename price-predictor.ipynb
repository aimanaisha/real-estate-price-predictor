```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True        
```

```python
df2[~df2['total_sqft'].apply(is_float)]
```

```python
def convert_to_num(x):
    a = x.split('-')
    if len(a)==2:
        return ((float(a[0])+float(a[1]))/2)
    try:
        return float(x)
    except:
        return None
```

```python
df3 = df2.copy()
df3['total_sqft'] = df.total_sqft.apply(convert_to_num)
df3.loc[122]
```

```python
df3['price_per_sqft'] = df3['price']*100000/df3['total_sqft']
```

```python
loc_stats = df3.location.value_counts().sort_values(ascending=False)
loc_stats_less_than_10 = loc_stats[loc_stats<10]
loc_stats_less_than_10
```

```python
df3.location = df3.location.apply(lambda x: 'other' if x in loc_stats_less_than_10 else x)
```
